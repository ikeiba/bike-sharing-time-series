Random Walk analysis


```python
#Imports 

import pandas as pd
import zipfile
import os
from pathlib import Path
```

```python

# Rutas
zip_path = "../../data.zip"
output_dir = Path("../../data")

# Creamos la carpeta de destino si no existe
output_dir.mkdir(parents=True, exist_ok=True)

# 1. Abrimos el archivo ZIP en modo lectura ('r')
with zipfile.ZipFile(zip_path, 'r') as zf:
    
    # zf.namelist() nos da una lista con los nombres de todo lo que hay dentro del ZIP
    for filename in zf.namelist():
        
        # 2. Filtramos para procesar solo los archivos .csv
        if filename.endswith('.csv'):
            print(f"Convirtiendo: {filename} ...")
            
            # 3. Abrimos el archivo específico de dentro del ZIP y lo leemos con Pandas
            with zf.open(filename) as f:
                df = pd.read_csv(f)
            
            # 4. Limpiamos el nombre para el nuevo archivo
            # Si filename es "carpeta_interna/ventas_2023.csv", nos quedamos solo con "ventas_2023"
            nombre_base = os.path.basename(filename).replace('.csv', '')
            ruta_parquet = output_dir / f"{nombre_base}.parquet"
            
            # 5. Guardamos en formato Parquet
            df.to_parquet(ruta_parquet, index=False)

print("¡Conversión completada!")
```
